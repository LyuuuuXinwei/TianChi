```python
def get_data_sliced(df):
    last = ''.join([df.columns[0].split('X')[0],'X'])
    sliced = 0
    slice_list = [0,]
    name_list = [last]
    for i in df.columns:
        label = ''.join([i.split('X')[0],'X'])
        sliced += 1
        if label != last:
            slice_list.append(sliced-1)
            last = label
            name_list.append(last)
    slice_list.append(len(df.columns))

    return name_list, slice_list
```

```python
def get_null_count(data):
    data_null_count = pd.DataFrame(index=data.index)
    data_null_count['null_count_all'] = data.isnull().sum(axis=1).values

    name_list, slice_list = get_data_sliced(data)
    for i in range(len(slice_list) - 1):
        data_null_count['null_count_{}'.format(name_list[i])] = data.iloc[:,
                                        slice_list[i]:(slice_list[i + 1] - 1)].isnull().sum(axis=1).values
    feature_null_count = data.isnull().sum(axis=0)

    return data_null_count, feature_null_count
```

```python
def get_low_variance_feature(df,threshold=0.02):
    delete_list=[]
    for i in df.columns:
        std=df[i].std()
        if std<threshold:
            delete_list.append(i)
    return delete_list
```

```python
#删除标准差低于0.02的特征
delete_list=get_low_variance_feature(data_num)
low_variance_feature=pd.DataFrame(index=data_num.index)
for i in delete_list:
    low_variance_feature[i]=data_num[i].values
data_num.drop(delete_list,axis=1,inplace=True)
low_variance_feature.to_csv(r'C:\Python\kaggle\@ON-工业AI\data\删除特征\low_variance_feature.csv')
```

```python
def get_majority(df,threshold=0.95):
    delete_list=[]
    for i in df.columns:
        cat=df[i].value_counts().values
        if cat.max()>(len(df)*threshold):
            delete_list.append(i)
    return delete_list
```

```python
#删除主成分大于0.98的特征
delete_list=get_majority(data_num)
data_num[delete_list].to_csv(r'C:\Python\kaggle\@ON-工业AI\data\删除特征\主成分大于0.98的特征.csv')
```

```python
# 删除表示时间的特征
def find_time_feature(data):
    df = data.fillna(0).copy()
    numerical_feature_median = pd.Series(np.zeros(len(df.columns)), index=df.columns)
    for i in df.columns:
        numerical_feature_median[i] = df[i].median()
    time_feature = numerical_feature_median[
        numerical_feature_median.astype('str').str.contains('2017') | numerical_feature_median.astype(
            'str').str.contains('2.017')]
    return time_feature.index
```

```python
#删除时间特征
time_feature = find_time_feature(data_num)
time=pd.DataFrame(index=data_num.index)
for i in time_feature:
    time[i]=data_num[i].values
data_num.drop(time_feature,axis=1,inplace=True)
time.to_csv(r'C:\Python\kaggle\@ON-工业AI\data\删除特征\time.csv')
```

```python
#合并高相似度的特征
def merge_similar_feature(data,threshold=0.98):
    df=data.copy()
    for i in df.columns:
        for j in df.columns:
            if i in df.columns:
                if j in df.columns:
                    if i != j:
                        if abs(df[i].corr(df[j])) > threshold:
                            if df[i].notnull().sum()>df[j].notnull().sum():
                                df.drop([j], axis=1, inplace=True)
                            else:
                                df.drop([i], axis=1, inplace=True)
    return df
```

```python
#删除互相相似的特征
data_num=merge_similar_feature(data_num)
```

```python
def feature_rank(df,label,mi_nighbors=[3,],na=0):
    x =df.loc[:499,:].fillna(na).copy()

    X=x.values
    Y=label[1].values
    
    tmp=pd.DataFrame(index=df.columns)
    
    f_test, _ = f_regression(X, Y)
    f_test /= np.max(f_test)
    tmp['f']=f_test
    tmp['p']=_
    
    corr=x.corrwith(label[1])
    tmp['corr']=corr
    
    for i in mi_nighbors:
        mi = mutual_info_regression(X, Y,n_neighbors=i,discrete_features=False)
        mi /= np.max(mi)
        tmp['mi_nb{0}'.format(i)]=mi
    
    return tmp

relation=feature_rank(data_num,y,mi_nighbors=[1,2,3,4,5])
```

```python
def tools_feature_relation(df,one_tool,mi_nighbors=[3,],na=0):
    x =df.fillna(na).copy()
    X=x.values
    Y=one_tool.values
    
    tmp=pd.DataFrame(index=x.columns)
    tmp.columns.name='{}'.format(one_tool.name)
    
    f_test, _ = f_regression(X, Y)
    f_test /= np.max(f_test)
    tmp['f']=f_test
    tmp['p']=_
    
    corr=x.corrwith(one_tool)
    tmp['corr']=corr
    
    for i in mi_nighbors:
        mi = mutual_info_regression(X, Y,n_neighbors=i,discrete_features=False)
        tmp['mi_{0}'.format(i)]=mi
        mi /= np.max(mi)
        tmp['mi_ratio{0}'.format(i)]=mi
    
    return tmp
```

```python
relation_tool=pd.DataFrame(index=data_num.columns)
for i in tools.columns:
    t=tools_feature_relation(data_num,tools[i],mi_nighbors=[1,2,3,4,5],na=-999)
    pd.concat([relation_tool,t],axis=1)
```

```python
def tools_related_feature(df,tools,corr_threshold=0.9,return_method='list'):
    dic_high_corr={}
    dic_grouped_median={}
    list_grouped_median=[]
    list_high_corr=[]
    for i in tools.columns:
        l_median=[]
        l_corr=[]
        for j in df.columns:
            tmp = pd.concat((tools[i], df[j]), axis=1)
            if len(tmp[j]) < (len(tmp[i]+10)):
                if tmp.groupby([i])[j].std().medain()==0:
                    l_median.append(j)
                    list_grouped_median.append(j)
                if abs(tmp[i].corr(tmp[j]))>corr_threshold:
                    l_corr.append(j)
                    list_high_corr.append(j)
        dic_grouped_median[i]=l_median
        dic_high_corr[i]=l_corr
    if return_method=='list':
        return list_grouped_median,list_high_corr
    else:
        return dic_grouped_median,dic_high_corr
```

```python
def delete_tools_related_feature(df,tools,corr_threshold=0.8,method='union'):
    data=df.copy()
    list_grouped_median,list_high_corr = tools_related_feature(df, tools,corr_threshold=corr_threshold,return_method='list')
    intersection = set(list_grouped_median) & set(list_high_corr)
    union = set(list_grouped_median) | set(list_high_corr)
    if method=='union':
        data.drop(union,axis=1,inplace=True)
        return union
    else:
        data.drop(intersection, axis=1, inplace=True)
        return intersection
```
